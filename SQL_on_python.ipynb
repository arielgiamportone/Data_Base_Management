```python
import mysql.connector
from mysql.connector import Error
```

Sobre una base de datos y para trabajar sobre ella, realizamos la conexión con el siguiente script donde se tienen en cuenta varios factores como el cursor, el commit, el tiempo de desconexion sin actividad (muy importante) y los errores:

```python
#Conexión del ejemplo de Demetrio:
try:     
    connection = mysql.connector.connect(       
        host = "tú_ip_host",        
        user = "tú_usuario",        
        password = "tú_contraseña",        
        database = "nombre_base_de_dato",        
        connection_timeout = 180     
        )    
    if connection.is_connected():        
        db_Info = connection.get_server_info()        
        print("Todo Correcto", db_Info)        
        cursor = connection.cursor()        
        connection.commit()
except Error as e:    
    print("Error while connecting to MySQL", e)
finally:    
    if (connection.is_connected()):        
        connection.close()        
        print("MySQL connection is closed")

```

```python
#Establecemos la conexión
connection = mysql.connector.connect(host = "tú_ip_host",        
        user = "tú_usuario",        
        password = "tú_contraseña",        
        database = "nombre_base_de_dato",        
        connection_timeout = 180)
#Conectamos el cursor, objeto importante para facilitar la ejecución de los comandos necesarios
cursor=connection.cursor()
```

Ejemplo de ejecucion de queries desde python en VSC:

```python
querie1 = """ SELECT * FROM PRODUCTO; """
cursor.execute(querie1)
cursor.fetchall()
```

```python
import pandas as pd
```

De la consulta en SQL obtendremos un  Data Frame con pandas (podria ser un CSV o lo que quisieramos, recordar la función de read):

```python
df1 = pd.read_sql(querie1, con= connection)
df1
```

Pasamos el data frame a csv y lo guardamos:

```python
df1.to_csv("frame_a_csv_con_pandas_importado_de_sql.csv", encoding= "latin1")
```

Configuramos para no tener limitaciones en lo que se muestra de la tabla:

Ver toda la informacion de las tablas:
